### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 4. Productos más y menos vendidos

```python
productos_mas_Vendidos = tienda['Producto'].value_counts().head(3)
productos_mas_Vendidos_tienda2 = tienda2['Producto'].value_counts().head(3)
productos_mas_Vendidos_tienda3 = tienda3['Producto'].value_counts().head(3)
productos_mas_Vendidos_tienda4 = tienda4['Producto'].value_counts().head(3)
```

```python
productos_menos_Vendidos = tienda['Producto'].value_counts().tail(3)
productos_menos_Vendidos_tienda2 = tienda2['Producto'].value_counts().tail(3)
productos_menos_Vendidos_tienda3 = tienda3['Producto'].value_counts().tail(3)
productos_menos_Vendidos_tienda4 = tienda4['Producto'].value_counts().tail(3)
```

```python
productos_combi = pd.concat([productos_mas_Vendidos, productos_menos_Vendidos])
productos_combi_tienda2 = pd.concat([productos_mas_Vendidos_tienda2, productos_menos_Vendidos_tienda2])
productos_combi_tienda3 = pd.concat([productos_mas_Vendidos_tienda3, productos_menos_Vendidos_tienda3])
productos_combi_tienda4 = pd.concat([productos_mas_Vendidos_tienda4, productos_menos_Vendidos_tienda4])
```

```python
productos_combi.plot(kind='bar',title="Tienda 1")
```

The final report recommends **Tienda 3** to Mr. Juan. This recommendation is justified by its highest average customer rating (4.05), competitive shipping costs, and a balanced performance profile that suggests a strong foundation for sustainable growth and customer satisfaction, despite not having the highest total revenues.

### Data Analysis Key Findings
*   **Total Revenues**: Tienda 1 generated the highest revenues at \$1,150,880,400, while Tienda 4 had the lowest at \$1,038,375,700.
*   **Sales by Category**: "Muebles" and "Electrónicos" were consistently the best-selling categories across all stores. Tienda 3 led in "Muebles" sales with 499 units, and Tienda 1 and Tienda 4 tied in "Electrónicos" with 451 units each. "Libros," "Instrumentos Musicales," and "Artículos para el hogar" were generally the lowest-selling categories.
*   **Average Ratings**: Tienda 3 achieved the highest average rating of 4.05, closely followed by Tienda 2 with 4.04. Tienda 1 had the lowest average rating at 3.98.
*   **Best and Worst-Selling Products**: The top and bottom-selling products varied significantly by store, indicating diverse demand or inventory specifics. For instance, "Microondas" was a top seller in Tienda 1 and Tienda 2, but a less sold product in Tienda 3.
*   **Average Shipping Costs**: Tienda 4 had the lowest average shipping cost at \$23,459.46, while Tienda 1 had the highest at \$26,018.61.

### Insights or Next Steps
*   Tienda 3 should leverage its high customer satisfaction (highest average rating) to further enhance its market position and explore strategies to boost sales in its less-performing categories like "Microondas," which are strong sellers elsewhere.
*   For Tienda 1, efforts should focus on improving customer satisfaction and optimizing shipping costs to become more competitive, as these factors could be hindering its overall profitability despite high revenues.
